```python
from pathlib import Path
import shutil
import random

# Ruta de tu dataset
dataset_path = Path("My First Project.yolov8")

train_img = dataset_path / "train" / "images"
train_lbl = dataset_path / "train" / "labels"

valid_img = dataset_path / "valid" / "images"
valid_lbl = dataset_path / "valid" / "labels"

# Crear carpetas valid si no existen
valid_img.mkdir(parents=True, exist_ok=True)
valid_lbl.mkdir(parents=True, exist_ok=True)

# Tomar todas las imágenes del train
image_files = list(train_img.glob("*"))
print("Imágenes encontradas en train:", len(image_files))

# Mezclar y separar 20% para valid
random.seed(42)
random.shuffle(image_files)

val_size = max(1, int(len(image_files) * 0.2))
val_images = image_files[:val_size]

print("Imágenes que irán a valid:", len(val_images))

# Mover imágenes y labels correspondientes
for img_path in val_images:
    label_path = train_lbl / f"{img_path.stem}.txt"
    
    shutil.move(str(img_path), str(valid_img / img_path.name))
    
    if label_path.exists():
        shutil.move(str(label_path), str(valid_lbl / label_path.name))

print("Split train/valid creado correctamente.")
```

```python
import cv2
import time
from ultralytics import YOLO

model_path = r"runs\detect\pokemon_detector\squirtle_yolov8n-2\weights\best.pt"
model = YOLO(model_path)

cap = cv2.VideoCapture(0)

if not cap.isOpened():
    raise RuntimeError("No se pudo abrir la webcam.")

prev_time = 0

while True:
    ret, frame = cap.read()
    if not ret:
        break

    start = time.time()

    results = model(frame, conf=0.25, iou=0.70, verbose=False)
    result = results[0]

    if result.boxes is not None and len(result.boxes) > 0:
        xyxy = result.boxes.xyxy.cpu().numpy()
        cls_ids = result.boxes.cls.cpu().numpy().astype(int)
        confs = result.boxes.conf.cpu().numpy()

        for box, cls_id, score in zip(xyxy, cls_ids, confs):
            x1, y1, x2, y2 = map(int, box)
            label = f"{model.names[cls_id]} {score:.2f}"

            cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 255), 2)
            cv2.putText(
                frame,
                label,
                (x1, max(20, y1 - 10)),
                cv2.FONT_HERSHEY_SIMPLEX,
                0.7,
                (0, 255, 255),
                2
            )

    end = time.time()
    fps = 1 / (end - start) if (end - start) > 0 else 0

    cv2.putText(
        frame,
        f"FPS: {fps:.2f}",
        (10, 30),
        cv2.FONT_HERSHEY_SIMPLEX,
        0.8,
        (255, 255, 255),
        2
    )

    cv2.imshow("Squirtle Detector", frame)

    key = cv2.waitKey(1) & 0xFF
    if key == 27 or key == ord("q"):
        break

cap.release()
cv2.destroyAllWindows()
```
